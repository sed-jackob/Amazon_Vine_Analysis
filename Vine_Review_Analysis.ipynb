```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
df.head()
```

```python
filtered_df = df[df['total_votes']>=20]
```

```python
filtered_df
```

```python
new_df = filtered_df[ (filtered_df['helpful_votes'] / filtered_df['total_votes']) > 0.5 ]
```

```python
new_df
```

```python
paid_review_df = new_df[ new_df['vine'] == 'Y']
```

```python
paid_review_df
```

```python
unpaid_review_df = new_df[ new_df['vine'] == 'N']
```

```python
unpaid_review_df
```

```python
paid_reviews_count = len(paid_review_df)
paid_reviews_5star = len(paid_review_df[ paid_review_df['star_rating'] == 5.0 ])
percent_paid_5star_reviews = paid_reviews_5star / paid_reviews_count
```

```python
unpaid_reviews_count = len(unpaid_review_df)
unpaid_reviews_5star = len(unpaid_review_df[ unpaid_review_df['star_rating'] == 5.0 ])
percent_unpaid_5star_reviews = unpaid_reviews_5star / unpaid_reviews_count
```
